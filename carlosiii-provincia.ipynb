```python
# Leemos el conjunto de datos por provincias
from pandas import read_csv
lectura = read_csv('casos_tecnica_provincia.csv', encoding='latin-1', sep = ',', index_col = 1)
```

```python
import pandas as pd
df = pd.DataFrame(lectura)
```

```python
# Vamos a eliminar del dataframe todas aquellas filas en las que los casos totales sean 0
# De esta forma contabilizamos para cada provincia el numero de casos a partir del inicio de la 
# pandemia en la propia provincia
df.dropna(axis = 0, inplace = True)
df_casos_mayor_cero = df[df['num_casos'] != 0]
```

```python
df_MU = df_casos_mayor_cero[df_casos_mayor_cero['provincia_iso'] == 'MU']
```

```python
df_MU.head(10)
```

```python
df_MU_casos = df_MU['num_casos']
# Visualización de los datos con un plot simple
from matplotlib import pyplot
df_MU_casos.plot()
# Rotamos 45º las fechas para que se vean correctamente
pyplot.xticks(rotation=90)
pyplot.show()
```

```python
from matplotlib.pyplot import legend
# Agrupamos por provincias y mostramos la serie dada por 'num_casos'
# df.groupby(by=grouping_columns)[columns_to_show].function()
df.groupby(['provincia_iso'])['num_casos'].plot()
# Pintamos la leyenda fuera de la zona grafica
legend(bbox_to_anchor=(1, 1.05), ncol = 4)
pyplot.xticks(rotation=90)
pyplot.show()
```
